```python
def jour(dat):
    dat=pd.to_datetime(dat)
    return dat
```

```python
data['date']=data['Date_of_Journey'].apply(jour)
```

```python
data.head()
```

```python
def day(x):
    res=x.day
    return res
```

```python
def month(x):
    res=x.month
    return res
```

```python
def year(x):
    res=x.year
    return res
```

```python
data['day']=data['date'].apply(day)
data['month']=data['date'].apply(month)
data['year']=data['date'].apply(year)
```

```python
data.head()
```

```python
data.drop(['Date_of_Journey','date'],axis=1,inplace=True)
```

```python
data.dtypes
```

```python
def change(dat):
    return pd.to_datetime(dat)
```

```python
data.head()
data.isna().sum()
data.dropna(inplace=True)
```

```python
data['arrival']=data['Arrival_Time'].apply(change)
data['dep']=data['Dep_Time'].apply(change)
```

```python
data.head()
```

```python
def close(dat):
    return dat.hour
def closed(dat):
    return dat.minute
```

```python
data['arrival_hour']=data['arrival'].apply(close)
data['arrival_minute']=data['arrival'].apply(closed)
```

```python
data['dep_hour']=data['dep'].apply(close)
data['dep_minute']=data['dep'].apply(closed)
```

```python
data.drop(['arrival','dep'],axis=1,inplace=True)
```

```python
data.head()
```

```python
duration=list(data['Duration'])
```

```python
for i in range(len(duration)):
    if len(duration[i].split(" "))==2:
        pass
    else:
        if 'h' in duration[i]:
            duration[i]=duration[i]+' 0m'
        else:
            duration[i]='0h '+duration[i]
```

```python
data['Duration']=duration
```

```python
data.head()
```

```python
def clear(dat):
    dat=dat.split(' ')[0][:-1]
    return dat
```

```python
def clears(dat):
    dat=dat.split(' ')[1][:-1]
    return dat
```

```python
data['duration_hour']=data['Duration'].apply(clear)
data['duration_minute']=data['Duration'].apply(clears)
```

```python
data.head()
```

```python
data.drop('Duration',axis=1,inplace=True)
```

```python
data.head()
```

```python
data.drop(['Arrival_Time','Dep_Time'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.dtypes
```

```python
def turn(dat):
    return int(dat)
```

```python
data['duration_hour']=data['duration_hour'].apply(turn)
data['duration_minute']=data['duration_minute'].apply(turn)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.3)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
estimator=[int(i) for i in np.linspace(100,1000,500)]
```

```python
features=['auto','sqrt']
```

```python
max_depth=[int(i) for i in np.linspace(5,100,45)]
```

```python
min_split=[5,10,15,20,25,100]
```

```python
random_grid={
    'n_estimators':estimator,
    'max_features':features,
'max_depth':max_depth,
    'min_samples_split':min_split
}
```

```python
random=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,cv=3,verbose=2,n_jobs=-1)
```

```python
random.fit(x_train,y_train)
```

```python
random.best_params_
```

```python
y_predict=random.predict(x_test)
```

```python
mean_absolute_error(y_test,y_predict)
```

```python
r2_score(y_test,y_predict)
```
